```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)
dataset = pd.read_csv(r"C:\Users\Able\Desktop\SMSSpam filter dataset\SMSSpamcollection.txt", sep = '\t', header = None)
dataset.columns = ['label', 'body_text']
dataset.head()
```

```python
print("Number of rows: {}" .format(len(dataset)))
print("Number of columns: {}" .format(len(dataset.columns)))
print("Number of spam: {}" .format(len(dataset[dataset['label'] == 'spam'])))
print("Number of ham: {}" .format(len(dataset[dataset['label'] == 'ham'])))
print("Number of missing values in label: {}" .format(dataset['label'].isnull().sum()))
print("Number of missing values in body_text: {}" .format(dataset['body_text'].isnull().sum()))
```

```python
import string
string.punctuation
def remove_punct(text):
    body_text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return body_text_nopunct
dataset['body_text_clean'] = dataset['body_text'].apply(lambda x: remove_punct(x))
dataset.head()
```

```python
import re
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens
dataset['body_text_tokenized'] = dataset['body_text_clean'].apply(lambda x: tokenize(x.lower()))
dataset.head()
```

```python
import nltk
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized):
    text = [word for word in tokenized if word not in stopword]
    return text
dataset['body_text_nostopword'] = dataset['body_text_tokenized'].apply(lambda x: remove_stopwords(x))
dataset.head()
```

```python
ps = nltk.PorterStemmer()
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text
dataset['body_text_stemmed'] = dataset['body_text_nostopword'].apply(lambda x: stemming(x))
dataset.head()
```

```python
wn = nltk.WordNetLemmatizer()
def lemmatizing(token_text):
    text = [wn.lemmatize(word) for word in token_text]
    return text
dataset['body_text_lemmatized'] = dataset['body_text_nostopword'].apply(lambda x: lemmatizing(x))
dataset.head(10)
```
